```python
missing_matrix = df.isnull().astype(int)
missing_corr = missing_matrix.corr()

print(missing_corr)
```

```python
df_train = df[df['workclass'].notna() & df['occupation'].notna()]
df_missing_workclass = df[df['workclass'].isna()]

features = ['education', 'educational-num', 'marital-status',
            'relationship', 'race', 'gender', 'hours-per-week']
df_encoded = df_train.copy()
label_encoders={}
for i in features +['workclass']+['occupation']:
    if df_encoded[i].dtype == 'object':
        le = LabelEncoder()
        df_encoded[i] = le.fit_transform(df_encoded[i])
        label_encoders[i] = le
```

```python
X_train = df_encoded[features]
y_train = df_encoded[['workclass', 'occupation']]
clf_workclass = RandomForestClassifier()
clf_workclass.fit(X_train, y_train)
```

```python
df_missing = df_missing_workclass.copy()
for col in features:
    if df_missing[col].dtype == 'object':
        df_missing[col] = label_encoders[col].transform(df_missing[col])
X_missing = df_missing[features]
y_pred = clf_workclass.predict(X_missing)
df_missing['workclass'] = label_encoders['workclass'].inverse_transform(y_pred[:,0])
df_missing['occupation'] = label_encoders['occupation'].inverse_transform(y_pred[:,1])
```

```python
df.loc[df['workclass'].isna(), 'workclass'] = df_missing['workclass']
df.loc[df['occupation'].isna(), 'occupation'] = df_missing['occupation']
```

```python
df = df.dropna(subset=['native-country'])

missing_values = df.isnull().sum()

print(missing_values)
```
